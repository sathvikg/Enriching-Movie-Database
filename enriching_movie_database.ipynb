```python
# IMDB Dataset
#Using SAX
from xml.sax import parse
from xml.sax.handler import ContentHandler

class imdb_Handler(ContentHandler):
  def __init__(self, total_movies):
    self.movie_key = 1
    self.total_movies = total_movies

  def startElement(self, name, attrs):
    if name == "movie":
      print("\n")
    print("<"+name+">",end="")

  def endElement(self, name):
    if name == "movie" and self.movie_key>=self.total_movies:
      print("</"+name+">",end="")
      raise Exception("")
    elif name == "movie" and self.movie_key<self.total_movies:
      self.movie_key+=1
      print("</"+name+">")
    else:
      print("</"+name+">")

  def characters(self, content):
      if content.strip() != "":
          print(repr(content),end="")

try:
  parse("/content/drive/MyDrive/DS-DNIT assignment/imdb-ds.xml", imdb_Handler(5))
except:
  pass

```
